```python
# Python Basics for Beginners

# 1. Variable Types
# Integer: Whole number without a decimal point
age = 25
print("Integer example:", age)
```

```python
# Float: Number with a decimal point
temperature = 98.6
print("Float example:", temperature)
```

```python
# String: A sequence of characters
name = "Alice"
print("String example:", name)
```

```python
# Boolean: True or False value
is_student = False
print("Boolean example:", is_student)
```

```python
# List: Ordered collection of items
fruits = ["apple", "banana", "cherry"]
print("List example:", fruits)
```

```python
# Tuple: Ordered, immutable collection of items
dimensions = (1920, 1080)
print("Tuple example:", dimensions)
```

```python
# Dictionary: Key-value pairs
student = {"name": "Bob", "age": 20}
print("Dictionary example:", student)
```

```python
# 2. Functions
# Defining a simple function
def greet(name):
    return "Hello, " + name + "!"
```

```python
# Using the function
greeting = greet("Alice")
print("Function example:", greeting)
```

```python
# 3. Conditional Statements
# If-else statement
age = 15
if age > 18:
    print("Adult")
else:
    print("Not an adult")
```

```python
# 4. Loops
# For loop: Iterating over a list
for i in fruits:
    print("Fruit:", i)
```

```python
# While loop: Repeating code as long as a condition is true
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

```python
# Importing Pandas Library
import pandas as pd
```

```python
# Basic Operations with Pandas

# Loading a Dataset
# Make sure to replace 'path/to/spotify_dataset.csv' with the actual path to the downloaded dataset
file_path = 'spotify-2023.csv'
# Try using a different encoding such as 'ISO-8859-1' if 'utf-8' doesn't work
try:
    spotify_data = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    spotify_data = pd.read_csv(file_path, encoding='ISO-8859-1')

print("Dataset loaded successfully.")
```

```python
# Displaying the first few rows of the dataset
print("\nFirst 10 rows of the dataset:")
print(spotify_data.head(10))
```

```python
# Displaying basic information about the dataset
print("\nDataset Information:")
spotify_data.info()
```

```python
# Descriptive statistics of the dataset
print("\nDescriptive Statistics:")
print(spotify_data.describe())
```

```python
# Accessing specific columns
# Example: Accessing the 'track_name' and 'artist(s)_name' columns
print("\nAccessing specific columns ('track_name' and 'artist(s)_name'):")
print(spotify_data[['track_name', 'artist(s)_name']].head())
```

```python
# Grouping and aggregating data
# Example: Finding the average danceability percentage per artist
print("\nAverage danceability percentage per artist:")
average_danceability = spotify_data.groupby('artist(s)_name')['danceability_%'].sum()
print(average_danceability.head())
```

```python
#NumPy is used for performing efficient array operations and numerical computations in Python.
# Import NumPy Library
import numpy as np

# Select only the numerical columns for analysis
numerical_columns = ['streams', 'danceability_%', 'energy_%']  # Replace with the actual numerical columns you need
numerical_data = spotify_data[numerical_columns].copy()

# Make sure all the data in 'numerical_data' is of numeric type
for column in numerical_columns:
    numerical_data[column] = pd.to_numeric(numerical_data[column], errors='coerce')

# Drop rows with NaN values that resulted from conversion errors
numerical_data = numerical_data.dropna()

# Now you can convert this to a NumPy array
numerical_array = numerical_data.to_numpy()
```

```python
# Perform NumPy operations
# Calculate the mean of the 'streams' column
streams_mean = np.mean(numerical_array[:, 0])  # Index 0 assumes 'streams' is the first column in the array
print("Average number of streams:", streams_mean)
```

```python
# Calculate the standard deviation of 'danceability_%'
danceability_std = np.std(numerical_array[:, 1])  # Index 1 assumes 'danceability_%' is the second column
print("Standard deviation of danceability:", danceability_std)
```

```python
# Calculate the median of 'energy_%'
energy_median = np.median(numerical_array[:, 2])  # Index 2 assumes 'energy_%' is the third column
print("Median energy percentage:", energy_median)
```

```python
import matplotlib.pyplot as plt
```

```python
# Scatter Plot of Danceability vs Energy
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(numerical_array[:, 1], numerical_array[:, 2], alpha=0.5)  # Plot danceability vs energy
plt.title('Danceability vs Energy')  # Set the title of the plot
plt.xlabel('Danceability (%)')  # Set the x-axis label
plt.ylabel('Energy (%)')  # Set the y-axis label
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```

```python
# Histogram of Streams
plt.figure(figsize=(10, 6))  # Set the figure size
plt.hist(numerical_array[:, 0], bins=50, color='blue', edgecolor='black', alpha=0.7)  # Create a histogram of streams
plt.title('Distribution of Streams')  # Set the title of the histogram
plt.xlabel('Number of Streams')  # Set the x-axis label
plt.ylabel('Frequency')  # Set the y-axis label
plt.grid(True)  # Show grid
plt.show()  # Display the histogram
```

```python
# Scatter Plot with Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='danceability_%', y='energy_%', data=spotify_data)
plt.title('Danceability vs Energy (Seaborn)')
plt.xlabel('Danceability (%)')
plt.ylabel('Energy (%)')
plt.show()
```

```python
# Ensure that 'streams' is a numeric column
spotify_data['streams'] = pd.to_numeric(spotify_data['streams'], errors='coerce')

# Drop any NaN values that arise from conversion errors
spotify_data = spotify_data.dropna(subset=['streams'])

# Now apply the log transformation with np.log1p
# np.log1p is used for log transformation to handle zero streams count as well
spotify_data['log_streams'] = np.log1p(spotify_data['streams'])

# Proceed with the histogram using the transformed data
plt.figure(figsize=(10, 6))
sns.histplot(spotify_data['log_streams'], bins=50, kde=True, color='green')
plt.title('Log Transformed Distribution of Streams (Seaborn)')
plt.xlabel('Log of Number of Streams')
plt.ylabel('Frequency')
plt.show()


```

```python
# Convert 'released_year' to numeric and create a new column for year intervals
spotify_data['released_year'] = pd.to_numeric(spotify_data['released_year'], errors='coerce')
spotify_data = spotify_data.dropna(subset=['released_year'])

# Define year intervals
year_bins = [2010, 2015, 2020, 2025]
year_labels = ['2010-2014', '2015-2019', '2020-2024']
spotify_data['year_interval'] = pd.cut(spotify_data['released_year'], bins=year_bins, labels=year_labels, right=False)

# Now, create the box plot with year intervals
plt.figure(figsize=(10, 6))
sns.boxplot(x='year_interval', y='streams', data=spotify_data)
plt.yscale('log')  # Apply log scale due to wide range of 'streams'
plt.title('Stream Distribution by Release Year Intervals (Seaborn)')
plt.xlabel('Release Year Intervals')
plt.ylabel('Log of Streams')
plt.show()
```
